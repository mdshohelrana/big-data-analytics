```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Load the dataset (replace with the correct path)
file_path = 'student-mat.csv'  # Replace with your file path
df = pd.read_csv(file_path, delimiter=';')

# Display the first few rows of the dataframe
print("First few rows of the dataset:")
print(df.head())

# Describe the type of features
print("\nData Types and Info:")
print(df.info())

print("\nStatistical Summary of Numerical Features:")
print(df.describe())

# Identify categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
print("\nCategorical Columns:", categorical_columns)

# Transform categorical variables
# Here we demonstrate both label encoding and one-hot encoding

# Label Encoding example
label_encoder = LabelEncoder()
for column in categorical_columns:
    # We apply label encoding only on columns that have 2 or fewer unique values
    if len(df[column].unique()) <= 2:
        df[column] = label_encoder.fit_transform(df[column])
        print(f"Label Encoded Column: {column}")

# One-Hot Encoding example
df = pd.get_dummies(df, columns=[col for col in categorical_columns if col not in df.columns])

# Display the transformed DataFrame
print("\nTransformed DataFrame:")
print(df.head())
```
